```python
count = dict()
for i in dataset5['tags']:
    #print(i)
    temp = i.split('>')
    temp.pop() # removing last element which is empty char
    for j in temp:
        count[j[1:].lower()] = count.get(j[1:].lower(), 0) + 1
```

```python
new_dataset = dataset5.copy()
print(new_dataset)
```

```python
duplicatePosts = dataset3[dataset3['link_type_id'] == 3]
print(duplicatePosts)
```

```python
new_dataset.rename(columns = {'id':'post_id'}, inplace = True)
```

```python
myData = pd.merge(duplicatePosts,new_dataset,on='post_id')
myData
```

```python
no_of_days_toClose = list()
```

```python
from datetime import datetime
date_format = "%Y-%m-%d"
```

```python
for i in myData.values:
    close = str(i[2])
    close = close.split('T')
    close = close[0]
    #print(close)
    create = str(i[10])
    create = create.split('T')
    create = create[0]
    #print(create)
    close = datetime.strptime(close,date_format)
    create = datetime.strptime(create,date_format)
    delta = close - create
    no_of_days_toClose.append(delta.days)
```

```python
no_of_days_toClose
```

```python
print(min(no_of_days_toClose))
print(max(no_of_days_toClose))
```

```python
myList = {'0-500':0, '501-1000':0, '1001-1500':0, '1501-2000':0, '2001-2500':0, '2501-3000':0, '3000+':0}
for i in no_of_days_toClose :
    if i<501:
        myList['0-500']+=1
    elif i < 1001:
        myList['501-1000']+=1
    elif i < 1501:
        myList['1001-1500']+=1
    elif i < 2001:
        myList['1501-2000']+=1
    elif i < 2501:
        myList['2001-2500']+=1
    elif i < 3001:
        myList['2501-3000']+=1
    else :
        myList['3000+']+=1
       
```

```python
print(myList)
```

```python
  
plt.figure(figsize = (10, 10))
plt.bar(myList.keys(), myList.values(),0.5)
plt.title('Time required to close duplicate questions')
plt.xlabel("No. Of Days")
plt.ylabel("Frequency Count of closed Questions")
```
