```python
count = dict()
for i in dataset5['tags']:
    #print(i)
    temp = i.split('>')
    temp.pop() # removing last element which is empty char
    for j in temp:
        count[j[1:].lower()] = count.get(j[1:].lower(), 0) + 1
```

```python
new_dataset = dataset5.copy()
print(new_dataset)
```

```python
duplicatePosts = dataset3[dataset3['link_type_id'] == 3]
print(duplicatePosts)
```

```python
new_dataset.rename(columns = {'id':'post_id'}, inplace = True)
```

```python
myData = pd.merge(duplicatePosts,new_dataset,on='post_id')
myData
```

```python
duplicate_tags_count = dict()
for i in myData['tags']:
    #print(i)
    temp = i.split('>')
    temp.pop() # removing last element which is empty char
    for j in temp:
        duplicate_tags_count[j[1:].lower()] = duplicate_tags_count.get(j[1:].lower(), 0) + 1
```

```python
print(duplicate_tags_count)
```

```python
percentage = dict()
for i,j in duplicate_tags_count.items():
    percentage[i] = (j/count[i])*100
print(percentage)
```
