```python
duplicatePosts = dataset3[dataset3['link_type_id'] == 3]
print(duplicatePosts)
```

```python
duplicatePosts_count = dict()
```

```python
print(duplicatePosts['creation_date'].head(1))
```

```python
duplicateQue_monthwise_count = dict()
for i in duplicatePosts['creation_date']:
    #print(i)
    temp = str(i)
    #print(temp)
    temp = temp.split('T')
    #print(temp)
    #print(temp[0])
    temp = temp[0].split('-')
    #print(temp)
    month_year = temp[1]+'-'+temp[0]
    #print(month_year)
    #print('\n')
    
    duplicateQue_monthwise_count[month_year] = duplicateQue_monthwise_count.get(month_year,0) + 1
```

```python
print(duplicateQue_monthwise_count)
```

```python
x_duplicateQue_monthwise_count = list()
y_duplicateQue_monthwise_count = list()

for i,j in duplicateQue_monthwise_count.items() :
    x_duplicateQue_monthwise_count.append(i)
    y_duplicateQue_monthwise_count.append(j)
print(x_duplicateQue_monthwise_count,y_duplicateQue_monthwise_count)

plt.figure(figsize = (25, 25))
plt.plot(x_duplicateQue_monthwise_count, y_duplicateQue_monthwise_count)
plt.xlabel("Months")
plt.ylabel("Frequency Count of Duplicate Questions")
plt.title("Duplicate questions asked in each month")
plt.xticks(rotation=90)
plt.show()  # show first chart
```

```python
# ratio w.r.t. previous month
ratio = []
for i in range(len(y_duplicateQue_monthwise_count)) :
    if(i == 0):
         ratio.append(y_duplicateQue_monthwise_count[0])
         continue
    ratio.append(int((y_duplicateQue_monthwise_count[i]/y_duplicateQue_monthwise_count[i-1])*100))
print(ratio)
    
ratio.pop(0)
```

```python
plt.figure(figsize = (25, 25))
plt.plot(x_duplicateQue_monthwise_count[1:], ratio)
plt.xlabel("Months")
plt.ylabel("Ratio")
plt.title("Ratio of duplicate questions asked in each month")
plt.xticks(rotation=90)
plt.show()  # show first chart
```
